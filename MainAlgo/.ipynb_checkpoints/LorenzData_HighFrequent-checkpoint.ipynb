```python
### Some initializations
import numpy as np
import random
import loli
import pickle
import copy
```

```python
# Generating coupled data

rs=1
np.random.seed(rs)
size=50000
X=np.zeros((size+1,6))
X[0,:]=[1,0.97,0.99,1,0.97,0.99]


for t in range(size):
    e=np.random.multivariate_normal(np.zeros(6),0.1*np.eye(6))
    x_new=[X[t,0]*0.9+0.1*X[t,1]+e[0],\
           0.25*X[t,0]-0.01*X[t,0]*X[t,2]+0.99*X[t,1]+0.001*X[t,4]**2+e[1],\
           0.01*X[t,0]*X[t,1]+0.9733*X[t,2]+e[2],\
           X[t,3]*0.9+0.1*X[t,4]+e[3],\
           X[t,3]*0.28-X[t,3]*X[t,5]*0.01+0.99*X[t,4]+0.0005*X[t,1]**2+e[4],\
           0.01*X[t,3]*X[t,4]+0.9733*X[t,5]+e[5]]
    X[t+1,:]=x_new
```

```python

np.random.seed(1)
E=500
n=20
d=6
runs=5
for tar in range(d):
    fpfn=np.zeros((runs,2))
    reportedS={}
    for j in range(runs):
        print('Run ',j+1,'out of ',runs,'target: ',tar)
        dataX=[]
        dataY=[]

        for e in range(E):
                p=j*n+e*n*runs
                dataX.append(X[p:p+n,:])
                dataY.append(X[p+1:p+1+n,tar])
        rs=np.random.randint(low=1,high=1000)
        plausibleS=loli.gauss(dataX,dataY,l=1,alpha=0.1,lam=0,B=500,rs=rs,intercept=False)
        if len(plausibleS)==0:
            supphat=set()
        else:
            supphat=set.intersection(*plausibleS)
        if tuple(supphat) in reportedS:
            reportedS[tuple(supphat)]+=1
        else:
            reportedS[tuple(supphat)]=1

    Lorenz={}
    Lorenz['results']=reportedS
    Lorenz['target']=tar
    Lorenz['runs']=runs
    Lorenz['samplesize']=n
    Lorenz['environments']=E

    with open('LorenzCoupledFixedRsE='+str(E)+'N'+str(n)+'tar'+str(tar)+'.pkl','wb') as f:
        pickle.dump(Lorenz, f)
```

```python
# What if we have more fine grained data?
# Generating coupled data: fine grained

rs=1
np.random.seed(rs)
size=100000
X_fine=np.zeros((size+1,6))
X_fine[0,:]=[1,0.97,0.99,-1,4,10]


for t in range(size):
    e=np.random.multivariate_normal(np.zeros(6),0.01*np.eye(6))
    e=np.zeros(6)
    e_obs=np.random.multivariate_normal(np.zeros(6),0.001*np.eye(6))
    x_new=[X_fine[t,0]*0.95+0.05*X_fine[t,1]+e[0],\
           0.125*X_fine[t,0]-0.005*X_fine[t,0]*X_fine[t,2]+0.995*X_fine[t,1]+0.0005*X_fine[t,4]**2+e[1],\
           0.005*X_fine[t,0]*X_fine[t,1]+0.98665*X_fine[t,2]+e[2],\
           X_fine[t,3]*0.95+0.05*X_fine[t,4]+e[3],\
           X_fine[t,3]*0.14-X_fine[t,3]*X_fine[t,5]*0.005+0.995*X_fine[t,4]+0.00025*X_fine[t,1]**2+e[4],\
           0.005*X_fine[t,3]*X_fine[t,4]+0.98665*X_fine[t,5]+e[5]]
    X_fine[t+1,:]=x_new+e_obs
```
