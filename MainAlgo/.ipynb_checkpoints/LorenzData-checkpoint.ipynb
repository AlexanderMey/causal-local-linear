```python
### Some initializations
import numpy as np
import math
import random
import copy
import matplotlib.pyplot as plt
import scipy
import scipy.stats
import lingam
import warnings
import itertools
import loli
from scipy.stats import binomtest
import pickle
from heapq import nlargest
from collections import Counter
from matplotlib.patches import Patch
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
#### Data creation of Lorenz systems that can couple and decouple via the couplings vector coup:

### The couplings vector. The first dim is the start end end of different coupling intervals.
### The second dim is the strength of the couplings vector, where 0 means no coupling. 
### Note that a too high coulingsvector will make the system diverge
coup=np.array([[0,10000,20000,30000,40000,60000],[5,5,5,5,5,5]])
###

T=int(coup[0,-1]) 
X=np.zeros((T+1,6))
X[0,:]=[1,0.97,0.99,2,4,-1]

                    
for r in range(len(coup[0,:])-1):
    alpha=5
    for t in range(int(coup[0,r]),int(coup[0,r+1])):
        if t<5:
            x_new=X[0,:]
        else:
            e=np.random.multivariate_normal(np.zeros(6),0.1*np.eye(6))
            x_new=[X[t,0]*0.9+0.1*X[t,1]+e[0],\
                   0.25*X[t,0]-0.01*X[t,0]*X[t,2]+0.99*X[t,1]+0.005*X[t,4]**2+e[1],\
                   0.01*X[t,0]*X[t,1]+0.9733*X[t,2]+e[2],\
                   X[t,3]*0.9+0.1*X[t,4]+e[3],\
                   X[t,3]*0.28-X[t,3]*X[t,5]*0.01+0.99*X[t,4]+0.001*X[t,1]**2+e[4],\
                   0.01*X[t,3]*X[t,4]+0.9733*X[t,5]+e[5]]



        X[t+1,:]=x_new
```

```python
E=500
n=35
d=6
tar=2
supp={0,1,2}
s=len(supp)
runs=1000
fpfn=np.zeros((runs,2))
reportedS={}
for j in range(runs):
    print('Run ',j+1,'out of ',runs)
    dataX=[]
    dataY=[]

    for e in range(E):
            p=np.random.randint(1,60000-n)
            dataX.append(X[p:p+n,:])
            dataY.append(X[p+1:p+1+n,tar])
    plausibleS=loli.gauss(dataX,dataY,l=1,alpha=0.1,lam=0,B=100)
    if len(plausibleS)==0:
        supphat=set()
    else:
        supphat=set.intersection(*plausibleS)
    if tuple(supphat) in reportedS:
        reportedS[tuple(supphat)]+=1
    else:
        reportedS[tuple(supphat)]=1

    
Lorenz={}
Lorenz['results']=reportedS
Lorenz['target']=tar
Lorenz['runs']=runs
Lorenz['samplesize']=n
Lorenz['environments']=E

with open('LorenzE'+str(E)+'N'+str(n)+'tar'+str(tar)+'.pkl','wb') as f:
    pickle.dump(Lorenz, f)
    
# l=[tuple(x) for x in fpfn]
# counts=Counter(l)
# c=np.array([[*key,val] for key,val in counts.items()])
# plt.scatter(c.T[0],c.T[1],10*c.T[2])
```
